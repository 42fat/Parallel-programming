       2.График ускорения(Ускорение = Время работы последовательного алгоритма / время работы параллельного алгоритма с заданным числом процессоров)

```python
acc_random = []
acc_sort = []
acc_unsort = []
for elem in time_random:
    acc_random.append(time_random[0]/elem)
for elem in time_sort:
    acc_sort.append(time_sort[0]/elem)
for elem in time_unsort:
    acc_unsort.append(time_unsort[0]/elem)

plt.figure(figsize=[16.,10.])
plt.plot(proc_count, acc_random, 'o-b', alpha=0.7, label="random",linewidth=2, ms=10)
plt.plot(proc_count, acc_sort, 'o-r', alpha=0.7, label="sort",linewidth=2, ms=10)
plt.plot(proc_count, acc_unsort, 'o-k', alpha=0.7, label="unsort",linewidth=2, ms=10)
plt.title("График ускорения")
plt.xlabel("Потоки")
plt.ylabel("Ускорение")
plt.legend()
plt.grid(True)
```

```python
eff_random = []
eff_sort = []
eff_unsort = []
for elem, i in zip(acc_random, proc_count):
    eff_random.append(elem/i)
for elem, i in zip(acc_sort, proc_count):
    eff_sort.append(elem/i)
for elem, i in zip(acc_unsort, proc_count):
    eff_unsort.append(elem/i)
plt.figure(figsize=[16.,10.])
plt.plot(proc_count, eff_random, 'o-b', alpha=0.7, label="random",linewidth=2, ms=10)
plt.plot(proc_count, eff_sort, 'o-r', alpha=0.7, label="sort",linewidth=2, ms=10)
plt.plot(proc_count, eff_unsort, 'o-k', alpha=0.7, label="unsort",linewidth=2, ms=10)
plt.title("График эффективности")
plt.xlabel("Потоки")
plt.ylabel("Эффективность")
plt.legend()
plt.grid(True)
```
