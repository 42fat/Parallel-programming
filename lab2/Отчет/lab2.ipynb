### Графики


    1.График Времени от количества потоков


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("report.txt", sep="\t", header=None)

array = data.to_numpy()
array = array.astype(float)
array[0][1]
a = array[0].shape

time_arr = []
proc_count = []
teor_time = []
i=0
for elem in array[0]:
    time_arr.append(elem)
for elem in array[0]:
    i=i+1
    proc_count.append(i)
for elem in proc_count:
    teor_time.append(array[0]/elem)
```

```python
plt.figure(figsize=[16.,6.])
plt.plot(proc_count, time_arr, 'o-b', alpha=0.7, label="time_per_proc",linewidth=2, ms=20)
##plt.plot(proc_count, teor_time, 'o-r', alpha=0.7, label="teor_time_per_proc",linewidth=2, ms=20)
plt.title("Времени от количества потоков")
plt.xlabel("Потоки")
plt.ylabel("Время")
plt.legend()
plt.grid(True)
```

       2.График ускорения(Ускорение = Время работы последовательного алгоритма / время работы параллельного алгоритма с заданным числом процессоров)
       

```python
acc_arr = []
for elem in time_arr:
    acc_arr.append(time_arr[0]/elem)
plt.figure(figsize=[16.,6.])
plt.plot(proc_count, acc_arr, 'o-b', alpha=0.7, label="acceleration",linewidth=2, ms=20)
plt.title("График ускорения")
plt.xlabel("Потоки")
plt.ylabel("Ускорение")
plt.legend()
plt.grid(True)
```

       3.График эффективности(Эффективность = Ускорение / число процессоров)

```python
eff_arr = []
for elem, i in zip(acc_arr, proc_count):
    eff_arr.append(elem/i)
plt.figure(figsize=[16.,6.])
plt.plot(proc_count, eff_arr, 'o-b', alpha=0.7, label="acceleration",linewidth=2, ms=20)
plt.title("График эффективности")
plt.xlabel("Потоки")
plt.ylabel("Эффективность")
plt.legend()
plt.grid(True)
```
